## 2.1 Stemming
stemming : the process of reducing a word to its root word

Example: There can be several words like 'actor','actress', 'acting'----->>>>Root word=Act

```python
port_stem = PorterStemmer()
```

```python
# def stemming(content):
#   stemmed_content = re.sub('[˄a-zA-Z]',' ',content)                 #sub--->>>substitute certain values, ˄ means excluding everything which is not present in a-z and A-Z
#   stemmed_content = stemmed_content.lower()
#   stemmed_content = stemmed_content.split()
#   stemmed_content = [port_stem.stem(word) for word in stemmed_content if not word in stopwords.words('english')]
#   stemmed_content = ' '.join(stemmed_content)
#   return stemmed_content 
```

```python
def stemming(content):
    stemmed_content = re.sub('[^a-zA-Z]',' ',content)
    stemmed_content = stemmed_content.lower()
    stemmed_content = stemmed_content.split()
    stemmed_content = [port_stem.stem(word) for word in stemmed_content if not word in stopwords.words('english')]
    stemmed_content = ' '.join(stemmed_content)
    return stemmed_content
```

```python
news_data.columns
```

```python
# Now apply above function i.e:  'stemming' to content column
news_data['content'] = news_data['content'].apply(stemming)
```

```python
print(news_data['content'])
```

Separating the data and Label

```python
X = news_data['content'].values     # Return a Numpy representation of the DataFrame.
Y = news_data['label'].values
```

```python
print(X)
print(Y)
```

##2.2 Converting the textual data to a numerical data

```python
vectorizer = TfidfVectorizer()    #Term frequecy inverse document frequency
```

```python
vectorizer.fit(X)
X = vectorizer.transform(X)
```

```python
print(X)
```

# 3.0 Splitting the data to training and test data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2, stratify=Y)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
print(Y.shape, Y_train.shape, Y_test.shape)
```

# 4.0 Training the Model: Logistic Regression

```python
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

# 005 Evaluation

```python
# Accuracy Score on the training data
X_train_prediction = model.predict(X_train)
print(X_train_prediction)
training_data_accuracy = accuracy_score(Y_train,X_train_prediction)
print("Accuracy of data is :", training_data_accuracy)
```

```python
# Accuracy Score on the test data
X_test_prediction = model.predict(X_test)
print(X_test_prediction)
test_data_accuracy = accuracy_score(Y_test,X_test_prediction)
print("Accuracy of data is :", test_data_accuracy)
```
